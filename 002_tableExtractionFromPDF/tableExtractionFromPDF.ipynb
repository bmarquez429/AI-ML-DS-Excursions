$\hspace{1in}$

###### **tabula-py**

##### Extract a table from a pdf document in Google Drive.

```python
tables = tabula.read_pdf(path + "sampleDocument.pdf", pages = 4)
tables[0]
```

$\hspace{1in}$

###### **camelot-py**

##### Extract a table from a pdf document in Google Drive.

```python
tables = camelot.read_pdf(path + "sampleDocument.pdf", pages = "4", flavor = "stream", table_areas = ["0, 570, 600, 90"])
tables[0].df
```
